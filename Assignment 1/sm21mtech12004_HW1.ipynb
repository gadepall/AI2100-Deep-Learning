# 3rd question

```python
def entropy(pmf):

  ent = 0
  for i in range(len(pmf)):
    if(pmf[i] == 0):
      continue
    ent += pmf[i] * np.log2(pmf[i]);

  
  return -ent
```

# 4a Normalized Histogram

```python
def norm_hist(imge):
 
  pixel_count  = np.zeros([256])
  height, breadth = imge.size 

  img = np.asarray(imge)

  for x_pixel in range(breadth):
    for y_pixel in range(height):
      value = int(img[x_pixel,y_pixel])
      pixel_count[value] += 1

  # Normalizing Histogram
  gray_norm = pixel_count / (height*breadth)

  return gray_norm
```

# 5th question

Joint Entropy


```python
def histogram_j(img_left,img_right):

  j_hist = np.zeros([256,256])    
  x = np.asarray(img_left) # 
  y = np.asarray(img_right) 

  x = x.flatten()
  y= y.flatten()

  for (im_l,im_r) in zip(x,y):
    j_hist[(im_l,im_r)] += 1

 
  j_hist = j_hist / j_hist.sum()

  return j_hist
```

```python
def conditional_pmf(joint_pmf, index : bool , value :int) :

  # index of conditioning random variable 
  """
  index = 0 ==> p(y|x = i)
  index = 1  ==> p(x|y = i)
  """

  # values of RV starts from 0
  if index == 0 :
    marginal_pmf_x  = joint_pmf[:,value].sum()
    if marginal_pmf_x > 0:
      return joint_pmf[:,value] / marginal_pmf_x 

  elif index == 1 : 
    marginal_pmf_y = joint_pmf[value,:].sum()
    if marginal_pmf_y > 0:
      return joint_pmf[value,:] / marginal_pmf_y

  else :
    print(f"Index of conditioning RV can either be 0 or 1")
```

```python
def conditional_entropy(joint_pmf , index : bool):

  
  """
  index = 0 ==> p(y|x = i)
  index = 1  ==> p(x|y = i)
  """
  c_entr = 0

  if index == 0 :
    
    for i in range(joint_pmf.shape[0]):
      c_pmf = conditional_pmf (joint_pmf,0,i)
      if c_pmf is not None:
        c_entr += -(joint_pmf[:,i] [joint_pmf[:,i] > 0 ] *  np.log2( c_pmf [c_pmf>0]) ).sum()

    return c_entr

  elif index == 1:

    
    for i in range(joint_pmf.shape[1]):
      c_pmf = conditional_pmf (joint_pmf,1,i) 
      if c_pmf is not None:
        c_entr += -(joint_pmf[i,:] [joint_pmf[i,:] > 0] * np.log2(c_pmf [c_pmf>0]) ).sum()

    return c_entr
```

# 7th Question

```python
def KL_divergence(p,q):
  k = (p[p>0]*np.log2(p [p>0] / q[q>0] )).sum()

  return k
```

```python
a = 0.4 # Ber(r) has P distribution
b = np.linspace(0,1,11)[1:-1]  # Q distribution has Ber(s)


Div_P_Q = (1-a)*( np.log2(1-a) - np.log2(1-b) ) + a*( np.log2(a) - np.log2(b) ) 
Div_Q_P = (1-b)*( np.log2(1-b) - np.log2(1-a) ) + b*( np.log2(b) - np.log2(a) ) 


plt.plot(b, Div_P_Q, color='r', label='D(P||Q)')
plt.plot(b, Div_Q_P, color='g', label='D(Q||P)')
  

plt.xlabel("Bernolli RV")
plt.ylabel("Divergence")
plt.title("K L divergence plots")
plt.legend()
plt.show()
```

```python
# P and Q be left and right images
left = Image.open('/content/left.png')
right = Image.open('/content/right.png')
p  = norm_hist(left)
q = norm_hist(right)

D_left_right = KL_divergence(p,q)
D_right_left = KL_divergence(q,p)

print(f" D(left||right) is {D_left_right}")
print(f"D(right||left) is {D_right_left}")
```

Observations:
The graph reaches a minimum value of zero at b = 0.4 in the preceding plot (KL divergence vs range of values in S). As a result, neither D(P||Q) nor D(Q||P) are negative graphs.
At b=0.4, the KL divergences are zero, and the value of an is set to 0.4. As a result, the KL divergences are equal for a = b.
